```python
df['전투력'] = 98
df
```

```python
df.drop(columns=['전투력'], inplace = True)
df
```

```python
df['재산'] = df['재산'].fillna(df['재산'].mean())
```

```python
df
```

```python
def add_cm(a):
    return str(a) + 'cm'

df['키'] = df['키'].apply(add_cm)
df
```

```python
def remove_cm(b):
    part = b.replace('cm', '')
    part_float = float(part)
    return part_float
df['키'] = df['키'].apply(remove_cm)
df
```

```python
df.groupby('국적')['나이'].mean()
```
