```python
missions = pd.DataFrame()
```

```python
rock_samples['Mission'].unique()
```

```python
missions['Mission'] = rock_samples['Mission'].unique()
missions
```

```python
sample_total_weight = rock_samples.groupby('Mission')['Weight (kg)'].sum()
```

```python
missions['Weight (kg)'] = sample_total_weight.values
missions
```

```python
missions.columns = ['Mission', 'Sample weight (kg)']
missions
```

```python
temp = missions['Sample weight (kg)'].diff()
temp
```

```python
missions['weight diff'] = temp
missions
```

```python
missions['weight diff'].fillna(value=0, inplace=True)
missions
```

```python
달모듈이름 = ['Eagle (LM-5)', 'Intrepid (LM-6)', 'Antares (LM-8)', 'Falcon (LM-10)', 'Orion (LM-11)', 'Challenger (LM-12)']
달모듈중량 = [15103, 15235, 15264, 16430, 16445, 16456]

missions['Lunar module (LM)'] = 달모듈이름
missions['LM mass (kg)'] = 달모듈중량
missions
```

```python
missions['LM mass diff'] = missions['LM mass (kg)'].diff()
missions
```

```python
missions['LM mass diff'].fillna(value=0, inplace=True)
missions
```

```python
명령모듈이름 = ['Columbia (CSM-107]', 'Yankee Clipper (CM-100])', 'Kitty Hawk (CM-110)', 'Endeavor (CM-112)', 'Casper (CM-113)', 'America (CM-114)']
명령모듈중량 = [5560, 5609, 5758, 5875, 5840, 5960]

missions['Command Module (CM)'] = 명령모듈이름
missions['CM Mass (kg)'] = 명령모듈중량
missions
```

```python
CM_diff = missions['CM Mass (kg)'].diff()
missions['CM mass diff'] = CM_diff
missions.fillna(value=0, inplace=True)
missions
```

```python
total = missions['LM mass (kg)'] + missions['CM Mass (kg)']
missions['Total weight (kg)'] = total
missions
```

```python
missions['Total weight diff'] = missions['Total weight (kg)'].diff()
missions['Total weight diff'].fillna(value=0 , inplace=True)
missions
```

```python
saturnVPayload = 43500
missions['Crewed area : Payload'] = missions['Total weight (kg)']/saturnVPayload
missions
```

```python
missions['Sample : Crewed area'] = missions['Sample weight (kg)']/missions['Total weight (kg)']
missions
```

```python
missions['Sample : Payload'] = missions['Sample weight (kg)']/saturnVPayload
missions
```

```python
artemis_crewedArea = 26520

artemis_dict = {'Mission' : ['artemis1', 'artemis1b', 'artemis2'],
                'Total weight (kg)' : [artemis_crewedArea, artemis_crewedArea, artemis_crewedArea],
                'Payload (kg)' : [26988, 37965, 42955]
               }

artemis_mission = pd.DataFrame(artemis_dict)
artemis_mission
```

```python
crewed_pay_ratio = missions['Crewed area : Payload'].mean()
sample_crewed_ratio = missions['Sample : Crewed area'].mean()
sample_pay_ratio = missions['Sample : Payload'].mean(0)

print(crewed_pay_ratio)
print(sample_crewed_ratio)
print(sample_pay_ratio)
```

```python
artemis_mission['Sample weight from total (kg)'] = artemis_mission['Total weight (kg)']*sample_crewed_ratio
artemis_mission
```

```python
artemis_mission['Sample weight from payload (kg)'] = artemis_mission['Payload (kg)'] * sample_pay_ratio
artemis_mission
```

```python
artemis_mission['Estimated sample weight (kg)'] = (artemis_mission['Sample weight from total (kg)'] + artemis_mission['Sample weight from payload (kg)'])/2
artemis_mission
```
