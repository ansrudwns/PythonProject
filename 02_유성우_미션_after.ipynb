```python
# 도시명을 입력하면 관측가능한 유성우에 대한 정보를 알려주는 함수 작성하기
    # 안내 메시지 초기화
    # 입력한 도시 정보가 존재하지 않으면 오류 메시지
def predict_best_meteor_shower_viewing(city):

    meteor_shower_string = ''
    if city not in cities['city'].values:
            meteor_shower_string = f'아쉽지만 {city}에서는 현재 유성우 예측이 어렵습니다.'
            return meteor_shower_string
        # 입력한 도시 정보가 존재하지 않으면 오류 메시지
        # 'Busan 에서는 별자리 관측이 불가능합니다!'

        #도시 위도 정보 불러오기
    latitude = cities.loc[cities['city']== city, 'latitude'].iloc[0]

        #해당 도시의 위도에서 관측가능한 별자리 조회
    constellations_list = constellations.loc[(constellations['latitudestart'] >= latitude) & (constellations['latitudeend']<= latitude), 'constellation'].tolist()
    
    # 도시별 관찰 가능한 별자리 iterate
        # 별자리와 가장 가까운 유성우 조회
    for i in constellations_list:
       
        # 유성우가 보일 때의 달의 위상 조회
        meteor_shower = meteor_showers.loc[meteor_showers['radiant'] == i, 'name'].iloc[0]

        # 유성우 관측이 가능한 시작일과 종료일
        meteor_showers_startdate = meteor_showers.loc[meteor_showers['radiant'] == i, 'startdate'].iloc[0]
        meteor_showers_enddate = meteor_showers.loc[meteor_showers['radiant'] == i, 'enddate'].iloc[0]
        
        # 달이 눈에 보이는 첫 날 조회
        moon_phase_list = moon_phases.loc[(moon_phases['date'] >= meteor_showers_startdate) & (moon_phases['date'] <= meteor_showers_enddate)]
        best_moon_date = moon_phase_list.loc[moon_phase_list['percentage'].idxmin()]['date']
        # 사용자에게 정보전달
        
        
        meteor_shower_string += f"\n{city}에서는 {meteor_shower} 관측이 가능합니다. \n {meteor_shower}를 보려면 {i} 자리를 향해서 {best_moon_date.to_pydatetime().strftime('%Y년 %m월 %d일')}에 보는 것을 추천드려요!"

    
    return meteor_shower_string
```
